```python
# Load ninea_timeregistrations (assuming you already have this)
query_timeregistrations = """
SELECT
    productionordernumber AS ProductionOrderID,
    zoneid               AS ZoneID,
    assemblystepid       AS AssemblystepID,
    CAST(startdate AS datetime2(0)) AS StartTime,
    CAST(enddate   AS datetime2(0)) AS EndTime,
    employeenumber       AS EmployeeNumber,
    salesordernumber     AS SalesOrderID,
    username             AS username,
    DATEDIFF(SECOND, startdate, enddate) AS DurationSeconds
FROM dbo.ninea_timeregistrations
WHERE productionordernumber IS NOT NULL
  AND startdate IS NOT NULL
  AND enddate   IS NOT NULL
  AND enddate >= startdate
"""

df_timeregistrations = pd.read_sql(query_timeregistrations, conn)
df_timeregistrations['StartTime'] = pd.to_datetime(df_timeregistrations['StartTime'])
df_timeregistrations['EndTime']   = pd.to_datetime(df_timeregistrations['EndTime'])

print(f"Time registrations loaded: {len(df_timeregistrations):,} records")
```

```python
# Load ninea_assemblystep
query_assemblystep = """
SELECT 
    assemblystepid                     AS AssemblystepID,
    articlenumber                      AS ArticleID,
    buildtime                          AS BuildTime,
    CAST(datemodified AS datetime2(0)) AS DateModified,
    producttype                        AS ProdctType,
    stepnumber                         AS StepNumber,
    zoneid                             AS ZoneID,
    zonenumber                         AS ZoneNumber,
    zoneuid                            AS ZoneUid
FROM ninea_assemblystep
"""
df_assemblystep = pd.read_sql(query_assemblystep, conn)
df_assemblystep['DateModified'] = pd.to_datetime(df_assemblystep['DateModified'])

print(f"Assembly steps loaded: {len(df_assemblystep):,} records")
```

### 1.2. Data preparation

```python
# Store initial count
initial_count = len(df_timeregistrations)

# Drop records missing critical identifiers: production order, zone, start or end time
df_timeregistrations = df_timeregistrations.dropna(subset=['ProductionOrderID', 'ZoneID', 'StartTime', 'EndTime'])

# Store final count
final_count = len(df_timeregistrations)

# Calculate and print number of removed records
removed_count = initial_count - final_count
print(f"🧹 Removed {removed_count:,} records during filtering")
print(f"✅ Remaining {final_count:,} time registration records")
print(f"   Date range: {df_timeregistrations['StartTime'].min()} to {df_timeregistrations['EndTime'].max()}")
```

#### Remove extremely long durations
Rationale: A small fraction of records have extremely long durations (hours or days) caused by forgotten stop actions. For example, 0.25% of steps exceed 8 hours, with some outliers up to ~217 days – clearly not actual assembly times. Moreover, if a production order contains an absurd outlier, we remove that entire order’s data to avoid any contamination in total time

```python
# Store initial count
initial_count = len(df_timeregistrations)

# Identify production orders containing any step > 8 hours (28800s)
outlier_orders = df_timeregistrations.loc[df_timeregistrations['DurationSeconds'] > 28800, 'ProductionOrderID'].unique()

# Remove all records for those production orders
df_timeregistrations = df_timeregistrations[~df_timeregistrations['ProductionOrderID'].isin(outlier_orders)].copy()

# Store final count
final_count = len(df_timeregistrations)

# Calculate and print number of removed records
removed_count = initial_count - final_count
print(f"🧹 Removed {removed_count:,} records during filtering")
print(f"✅ Remaining {final_count:,} time registration records")
print(f"   Date range: {df_timeregistrations['StartTime'].min()} to {df_timeregistrations['EndTime'].max()}")
```

#### Correct overlapping timestamps
These overlaps indicate logging errors (e.g. an operator started a new step before stopping the previous one) and would cause double-counting if we simply sum durations. To fix this, we enforce chronological consistency within each production order: if a step’s start overlaps a previous step’s ongoing time, we adjust the earlier step’s end time to match the overlap start 

```python
# Store initial count
initial_count = len(df_timeregistrations)

# Sort by production order and start time
df_timeregistrations.sort_values(['ProductionOrderID', 'StartTime'], inplace=True)

# Identify overlapping sequential steps within the same order
mask_overlap = (
    (df_timeregistrations['ProductionOrderID'] == df_timeregistrations['ProductionOrderID'].shift()) & 
    (df_timeregistrations['StartTime'].shift() < df_timeregistrations['EndTime'].shift()) & 
    (df_timeregistrations['StartTime'] < df_timeregistrations['EndTime'].shift()) 
)
# For each overlap, truncate the previous step's end time to this step's start time
df_timeregistrations.loc[mask_overlap.shift(-1, fill_value=False), 'EndTime'] = df_timeregistrations['StartTime']

# Recompute durations in seconds after fixing overlaps
df_timeregistrations['DurationSeconds'] = (df_timeregistrations['EndTime'] - df_timeregistrations['StartTime']).dt.total_seconds()

# Store final count
final_count = len(df_timeregistrations)

# Calculate and print number of removed records
removed_count = initial_count - final_count
print(f"🧹 Removed {removed_count:,} records during filtering")
print(f"✅ Remaining {final_count:,} time registration records")
print(f"   Date range: {df_timeregistrations['StartTime'].min()} to {df_timeregistrations['EndTime'].max()}")
```
