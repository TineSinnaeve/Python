# Data Understanding – Assembly Time Dataset
CRISP‑DM Data Understanding notebook generated from SQL queries.
Includes distributions, statistics, outlier detection, overlap detection, and Pareto charts.

## 1. Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

## 2. Load Data
This assumes `df_timeregistrations` is already loaded from SQL.

```python
df = df_timeregistrations.copy()

df["StartTime"] = pd.to_datetime(df["StartTime"])
df["EndTime"] = pd.to_datetime(df["EndTime"])

df["DurationSeconds"] = (df["EndTime"] - df["StartTime"]).dt.total_seconds()
```

## 3. Basic Data Profiling

```python
print("Total records:", len(df))
print("Distinct orders:", df["ProductionOrderID"].nunique())
print("Distinct zones:", df["ZoneID"].nunique())
print("Distinct steps:", df["StepID"].nunique())
```

```python
steps_per_order = df.groupby("ProductionOrderID").size()
steps_per_order.describe()
```

## 4. Steps per Order Distribution + Pareto

```python
bins = [0,10,50,100,200,500,1000,np.inf]
labels = [
    "<10 steps","10-49 steps","50-99 steps",
    "100-199 steps","200-499 steps","500-999 steps",
    ">=1000 steps"
]

bucket = pd.cut(steps_per_order, bins=bins, labels=labels)
counts = bucket.value_counts().sort_index()

pareto = counts.sort_values(ascending=False)
cumperc = pareto.cumsum() / pareto.sum() * 100

fig, ax1 = plt.subplots()

pareto.plot(kind="bar", ax=ax1)
ax2 = ax1.twinx()
cumperc.plot(ax=ax2, marker="o")

ax1.set_ylabel("Frequency")
ax2.set_ylabel("Cumulative %")
plt.title("Pareto Chart – Steps per Order")
plt.show()
```

## 5. Duration Statistics

```python
df["DurationSeconds"].describe()
```

```python
df["DurationSeconds"].quantile([
    0.01,0.05,0.10,0.25,0.50,0.75,0.90,0.95,0.99
])
```

## 6. Duration Distribution Histogram

```python
plt.figure()
plt.hist(df["DurationSeconds"], bins=100)
plt.xlabel("Duration (seconds)")
plt.ylabel("Frequency")
plt.title("Step Duration Distribution")
plt.show()
```

## 7. Pareto Chart – Duration Buckets

```python
bins = [
0,1,5,10,30,60,120,300,600,900,1800,3600,7200,
14400,28800,86400,604800,np.inf]

labels = [
"<1s","1-4s","5-9s","10-29s","30-59s",
"1-2m","2-5m","5-10m","10-15m",
"15-30m","30-60m","1-2h","2-4h",
"4-8h","8-24h","1-7d",">1w"]

duration_bucket = pd.cut(df["DurationSeconds"], bins=bins, labels=labels)

counts = duration_bucket.value_counts()

pareto = counts.sort_values(ascending=False)
cumperc = pareto.cumsum() / pareto.sum() * 100

fig, ax1 = plt.subplots()

pareto.plot(kind="bar", ax=ax1)
ax2 = ax1.twinx()
cumperc.plot(ax=ax2, marker="o")

ax1.set_ylabel("Frequency")
ax2.set_ylabel("Cumulative %")
plt.title("Pareto Chart – Step Duration Buckets")
plt.show()
```

## 8. Short Duration Analysis

```python
thresholds = [1,2,5,10,30]

results = []

for t in thresholds:
    count = (df["DurationSeconds"] < t).sum()
    pct = count / len(df) * 100
    results.append((t,count,pct))

pd.DataFrame(results, columns=["Threshold_sec","Count","Percentage"])
```

## 9. Long Duration Analysis

```python
thresholds = [3600,14400,28800,86400,604800]

results = []

for t in thresholds:
    count = (df["DurationSeconds"] > t).sum()
    pct = count / len(df) * 100
    results.append((t,count,pct))

pd.DataFrame(results, columns=["Threshold_sec","Count","Percentage"])
```

## 10. IQR Outlier Detection

```python
Q1 = df["DurationSeconds"].quantile(0.25)
Q3 = df["DurationSeconds"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Lower bound:", lower)
print("Upper bound:", upper)

low_outliers = (df["DurationSeconds"] < lower).sum()
high_outliers = (df["DurationSeconds"] > upper).sum()

print("Low outliers:", low_outliers)
print("High outliers:", high_outliers)
```

## 11. Overlap Detection

```python
df_sorted = df.sort_values(["ProductionOrderID","StartTime"])

df_sorted["NextStart"] = df_sorted.groupby("ProductionOrderID")["StartTime"].shift(-1)

df_sorted["Overlap"] = df_sorted["EndTime"] > df_sorted["NextStart"]

df_sorted["Overlap"].sum()
```

## 12. Gap Analysis

```python
df_sorted["GapSeconds"] = (
    df_sorted["NextStart"] - df_sorted["EndTime"]
).dt.total_seconds()

df_sorted["GapSeconds"].describe()
```

## 13. Zone Level Aggregation

```python
zone_order = (
    df.groupby(["ProductionOrderID","ZoneID"])
    .agg(
        num_steps=("StepID","count"),
        zone_start=("StartTime","min"),
        zone_end=("EndTime","max"),
        total_zone_duration=("DurationSeconds","sum"),
        avg_step_duration=("DurationSeconds","mean")
    )
    .reset_index()
)

zone_order.head()
```

## 14. Zone Time Pareto

```python
zone_totals = df.groupby("ZoneID")["DurationSeconds"].sum()

pareto = zone_totals.sort_values(ascending=False)
cumperc = pareto.cumsum() / pareto.sum() * 100

fig, ax1 = plt.subplots()

pareto.plot(kind="bar", ax=ax1)
ax2 = ax1.twinx()
cumperc.plot(ax=ax2, marker="o")

ax1.set_ylabel("Total Time (seconds)")
ax2.set_ylabel("Cumulative %")
plt.title("Pareto Chart – Time by Zone")
plt.show()
```

## 15. Total Assembly Time per Order

```python
order_times = df.groupby("ProductionOrderID").agg(
    total_steps=("StepID","count"),
    num_zones=("ZoneID","nunique"),
    first_step=("StartTime","min"),
    last_step=("EndTime","max"),
    total_assembly_time=("DurationSeconds","sum")
)

order_times["elapsed_time"] = (
    order_times["last_step"] - order_times["first_step"]
).dt.total_seconds()

order_times.head()
```

## 16. Assembly Time Distribution

```python
plt.figure()
plt.hist(order_times["total_assembly_time"]/3600, bins=50)
plt.xlabel("Assembly Time (hours)")
plt.ylabel("Orders")
plt.title("Distribution of Assembly Times")
plt.show()
```

## 17. Temporal Analysis

```python
df["Year"] = df["StartTime"].dt.year
df["Month"] = df["StartTime"].dt.month

df.groupby(["Year","Month"]).agg(
    num_registrations=("StepID","count"),
    num_orders=("ProductionOrderID","nunique"),
    avg_duration=("DurationSeconds","mean")
)
```
