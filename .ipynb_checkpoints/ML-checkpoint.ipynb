## 1. Database Connection and Data Loading

Connecting to Azure SQL Database containing Tobroco's ThingWorx shopfloor data.

```python
# Database connection string
conn_str = (
    "Driver={ODBC Driver 17 for SQL Server};"
    "Server=tobroco-ai-db.database.windows.net;"
    "Database=thingworx_data;"
    "UID=sqladmin;"
    "PWD=>Qy:NhMV[F1DX2h'!%;"
)

# Establish connection
conn = pyodbc.connect(conn_str)
print("✅ Connected to Azure SQL Database")
```

```python
# Load time registrations with data quality filters
query_time = """
SELECT
    productionordernumber AS ProductionOrderID,
    zoneid               AS ZoneID,
    assemblystepid       AS StepID,
    startdate            AS StartTime,
    enddate              AS EndTime,
    DATEDIFF(SECOND, startdate, enddate) AS DurationSeconds
FROM dbo.ninea_timeregistrations
WHERE productionordernumber IS NOT NULL
  AND startdate IS NOT NULL
  AND enddate   IS NOT NULL
  AND enddate >= startdate
"""

df = pd.read_sql(query_time, conn)
df['StartTime'] = pd.to_datetime(df['StartTime'])
df['EndTime'] = pd.to_datetime(df['EndTime'])

print(f"✅ Loaded {len(df):,} time registration records")
print(f"   Date range: {df['StartTime'].min()} to {df['EndTime'].max()}")
df.head()
```

```python
# Store initial count
initial_count = len(df)

# Drop records missing critical identifiers: production order, zone, start or end time
df = df.dropna(subset=['ProductionOrderID', 'ZoneID', 'StartTime', 'EndTime'])

# Store final count
final_count = len(df)

# Calculate and print number of removed records
removed_count = initial_count - final_count
print(f"🧹 Removed {removed_count:,} records during filtering")
print(f"✅ Remaining {final_count:,} time registration records")
print(f"   Date range: {df['StartTime'].min()} to {df['EndTime'].max()}")
```

#### Remove extremely long durations
Rationale: A small fraction of records have extremely long durations (hours or days) caused by forgotten stop actions. For example, 0.25% of steps exceed 8 hours, with some outliers up to ~217 days – clearly not actual assembly times. Moreover, if a production order contains an absurd outlier, we remove that entire order’s data to avoid any contamination in total time

```python
# Store initial count
initial_count = len(df)

# Identify production orders containing any step > 8 hours (28800s)
outlier_orders = df.loc[df['DurationSeconds'] > 28800, 'ProductionOrderID'].unique()

# Remove all records for those production orders
df = df[~df['ProductionOrderID'].isin(outlier_orders)].copy()

# Store final count
final_count = len(df)

# Calculate and print number of removed records
removed_count = initial_count - final_count
print(f"🧹 Removed {removed_count:,} records during filtering")
print(f"✅ Remaining {final_count:,} time registration records")
print(f"   Date range: {df['StartTime'].min()} to {df['EndTime'].max()}")
```

#### Correct overlapping timestamps
These overlaps indicate logging errors (e.g. an operator started a new step before stopping the previous one) and would cause double-counting if we simply sum durations. To fix this, we enforce chronological consistency within each production order: if a step’s start overlaps a previous step’s ongoing time, we adjust the earlier step’s end time to match the overlap start 

```python
# Store initial count
initial_count = len(df)

# Sort by production order and start time
df.sort_values(['ProductionOrderID', 'StartTime'], inplace=True)

# Identify overlapping sequential steps within the same order
mask_overlap = (
    (df['ProductionOrderID'] == df['ProductionOrderID'].shift()) & 
    (df['StartTime'].shift() < df['EndTime'].shift()) & 
    (df['StartTime'] < df['EndTime'].shift()) 
)
# For each overlap, truncate the previous step's end time to this step's start time
df.loc[mask_overlap.shift(-1, fill_value=False), 'EndTime'] = df['StartTime']

# Recompute durations in seconds after fixing overlaps
df['DurationSeconds'] = (df['EndTime'] - df['StartTime']).dt.total_seconds()

# Store final count
final_count = len(df)

# Calculate and print number of removed records
removed_count = initial_count - final_count
print(f"🧹 Removed {removed_count:,} records during filtering")
print(f"✅ Remaining {final_count:,} time registration records")
print(f"   Date range: {df['StartTime'].min()} to {df['EndTime'].max()}")
```

#### Aggregate Data at the Zone Level
Rationale: Given the persistent noise at step level, we switch to zone-level aggregation for modeling. Summing durations of all steps in a zone for each production order yields a more reliable measure of work per zone

```python
# Aggregate step durations and number of steps to zone-level per production order
zone_df = (
    df.groupby(['ProductionOrderID', 'ZoneID'], as_index=False)
      .agg(
          ZoneDurationSec =('DurationSeconds', 'sum'),
          NumberOfSteps=('StepID', 'nunique')  # Count unique steps
      )
)
```

```python
zone_df.head(10)
```

```python
# Calculate total assembly time per order (target variable)
total_time_df = zone_df.groupby('ProductionOrderID', as_index=False)['ZoneDurationSec'].sum()
total_time_df.rename(columns={'ZoneDurationSec': 'TotalAssemblyTimeSec'}, inplace=True)
```

```python
total_time_df.head(10)
```

#### Address Skewness with Transformation
After this transformation, extremely long times are pulled closer to the center. For instance, an 8-hour duration (28800s) becomes ~10.27 in log-scale, and 30-second durations become ~3.4, instead of 28800 vs 30. This should help the model train more effectively on a wide range of assembly times.

```python
zone_df['logZoneDurationSec'] = np.log1p(zone_df['ZoneDurationSec'])
total_time_df['logTotalAssemblyTimeSec'] = np.log1p(total_time_df['TotalAssemblyTimeSec'])
```
